```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs

def movie_ranking(date, kind):

    # ìˆ˜ì§‘í•  url ê°€ì ¸ì˜¤ê¸°
    url = f"https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel={kind}&tg=0&date={date}"
    
    # requestsë¥¼ í†µí•œ HTTP ìš”ì²­, bsë¡œ html ìš”ì†Œ íŒŒì‹±
    response = requests.get(url, headers={"user-agent": "Mozilla/5.0"})
    html = bs(response.text)
    
    # í•„ìš”í•œ ì •ë³´ì¸ ì˜í™”ì œëª©, í‰ì  ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    movie_title = html.select(".title > div > a")
    movie_point = html.select(".point")
    
    # bs4.element.ResultSet íƒ€ì…ì„ list íƒ€ì…ìœ¼ë¡œ ë³€ê²½
    title_list = []
    point_list = []
    rank_list = []

    for i in range(len(movie_title)):
        title_list.append(movie_title[i].contents[0])
        point_list.append(movie_point[i].contents[0])
        rank_list.append(i+1)
    
    # list íƒ€ì…ì„ DataFrame ìœ¼ë¡œ ë³€ê²½
    df = pd.DataFrame(point_list, columns = ['í‰ì '])
    df.insert(0, "ì˜í™”ì œëª©", title_list, True)
    df = df.set_index([pd.Index(rank_list)])
    
    return df

movie_ranking(20220505, 'cur') 
```
