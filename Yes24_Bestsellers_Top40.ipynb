# Pandas로 가져오기

```python
import pandas as pd
import numpy as np
# import requests
```

## 1. 베스트셀러 페이지 url 가져오기

```python
page_no = 1

url = f"http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber={page_no}"
```

```python
table = pd.read_html(url)
```

### 2. table[0]의 열 2번의 데이터가 베스트셀러 도서 리스트

```python
table[0]
```

```python
best = table[0][2]
```

```python
best
```

```python
type(best)
```

### 3. best 에 있는 NaN 값 제거

```python
best = best.dropna(axis=0)
```

```python
best
```

### 4. index 값 리셋하기

```python
best = best.reset_index(drop=True)
```

### 5. 데이터 분할 (split)

```python
books = best.str.split('|').str[0]
```

```python
# 책 정보에서 [도서], (개정판) 등의 정보 제거하기
book_list = []

for i, book in enumerate(books):
    book = book[5:].split(" : ")[0]
    book = book.split("(")[0]
    book_list.append(book)
```

```python
book_list
```

```python
publishers = best.str.split('|').str[1]
publisher_list = []
```

```python
for j, publisher in enumerate(publishers):
    publisher_list.append(publisher.strip()) # 양쪽 공백 제거
```

```python
publisher_list
```

### 6. 각 정보를 dataframe으로 만들기

```python
df = pd.DataFrame(list(zip(book_list, publisher_list)), columns=['도서', '출판사'])
```

```python
df
```

### 7. 순위 정보 넣기

```python
rank_list = []

for i in range(1, len(df)+1):
    rank_list.append(i)
```

```python
df.insert(0, '순위', rank_list)
```

```python
df = df.set_index(keys=['순위'], drop=True)
```
